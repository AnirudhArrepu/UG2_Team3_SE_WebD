```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch

model_name = "meta-llama/Llama-3.2-3B"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True, bnb_4bit_use_double_quant=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.add_special_tokens({'pad_token': '<PAD>'})

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config
)
```

```python
!pip install pipeline
```

```python
import transformers
from transformers import pipeline

pline = pipeline(
    model=model,
    tokenizer=tokenizer,
    return_full_text=True,
    task='text-generation',
    max_new_tokens=512,
    temperature=0.7,
    top_p=0.95,
    repetition_penalty=1.15
    # use_cache=True
)
```

```python
!pip install PyPDF2
!pip install langchain-community
```

```python
from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=pline)
```

```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def generate_embeddings(text):
  return embedder.encode(text, convert_to_tensor=True)
```

```python
type(text)
len(text)
```

```python
import chromadb

client = chromadb.Client()

collection = client.create_collection("doc")

for i, chunk in enumerate(text):
    embedding = generate_embeddings(chunk).tolist()
    collection.add(
        documents=[chunk],
        embeddings=[embedding],
        ids=[f"{i}"]
    )
    # print(f"Added document {i}: {chunk[:3]}... with embedding: {embedding[:5]}")  # Print the first 30 chars and first 5 embedding values
```

```python
def get_related_query(query):
  embeds = generate_embeddings(query)
  # print(query)
  # print(embeds)
  results = collection.query(
      # query_embeddings = [embeds.tolist()],
      query_texts = query,
      n_results = 3
  )
  # print(results['documents'])
  return results['documents']
```

```python
def get_answer(query, context):
    print("------------------------")
    # context_text = " ".join(context)  # Join if context is a list
    prompt = f"Context: {context}\n\nQuery: {query}\n\nAnswer:"
    # input = tokenizer(prompt, return_tensors='pt')
    # output = model.generate(**input, max_length=300)
    # return tokenizer.decode(output[0], skip_special_tokens=True)
    return llm.predict(prompt)
```

```python
query = "what is the vision of aravaind eye care"
context = (get_related_query(query))
answer = get_answer(query, context)
answer
```

```python
def answer_query(query):
  context = (get_related_query(query))
  answer = get_answer(query, context)
  return answer
```
